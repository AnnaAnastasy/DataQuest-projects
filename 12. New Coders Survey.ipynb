# Finding the Best Markets to Advertise In
The project is made for an e-learning company that offers courses on programming. Most of the courses are on web and mobile development, but there are also many other domains, like data science, game development, etc.
the goal of this project is to find out the two best markets to invest money in advertisement.

To reach this goal, I could organize surveys for a couple of different markets to find out which would the best choices for advertising. This is very costly, however, and it's a good call to explore cheaper options first.

The cheaper option is to search existing data that might be relevant for the purpose of the project. One good candidate is the data from [freeCodeCamp's 2017 New Coder Survey](https://www.freecodecamp.org/news/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969/). [freeCodeCamp](https://www.freecodecamp.org/) is a free e-learning platform that offers courses on web development. Because they run a popular Medium publication (over 400,000 followers), their survey attracted new coders with varying interests (not only web development), which is ideal for the purpose of the analysis.

The survey data is publicly available in [this GitHub repository](https://github.com/freeCodeCamp/2017-new-coder-survey).

## New Coders - Locations and Densities
I will start analysis with finding out where the new coders are located, and what are the densities (how many coderes there are) for each location.

The data set provides information about the location of each participant at a country level. The `CountryCitizen` variable describes the country of origin for each participant, and the `CountryLive` variable describes what country each participants lives in (which may be different than the origin country).

I will work with the `CountryLive` variable because I am interested where people actually live at the moment when the ads are run.

Because the data set provides information at a country level, each country can be seen as an individual market. This means that the ***goal of the project*** can be reframed as ***finding the two best countries to advertise in***.

One indicator of a good market is the number of potential customers — the more potential customers in a market, the better.

First of all, to make sure we are working with a representative sample, I will drop all the rows where participants didn't answer what role they are interested in:

```python
survey = survey[survey['JobRoleInterest'].notnull()].copy()
survey.shape
```

#### Generating frequency tables for the `CountryLive` variable:

```python
# absolute frequencies
survey['CountryLive'].value_counts()[:20]
```

```python
# relative frequencies
survey['CountryLive'].value_counts(normalize = True)[:20] * 100
```

We can see that 45% of the market is located in the United States, 7.7% - in India, 4.6% - in United Kingdom and 3.8% - in Canada. These are the top countries where new coders live, but I still need to go more in depth with the analysis before taking decision.

## Spending Money for Learning
The next step is to figure out how much money new coders are actually willing to spend on learning. Advertising within markets where most people are only willing to learn for free is extremely unlikely to be profitable.

It also seems like a good idea to narrow down the analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

- These are the countries having the highest absolute frequencies in the sample.
- The courses of our e-learning company are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our ads.

```python
# creating new column: money spent by student per month
survey['MonthsProgramming'] = survey['MonthsProgramming'].apply(
    lambda x: 1 if x == 0 else x
    ) #to avoid diving by 0

survey['MoneyPerMonth'] = survey['MoneyForLearning'] / survey['MonthsProgramming']
survey['MoneyPerMonth'][:5]
```

```python
# how many null values are in new column
survey['MoneyPerMonth'].isnull().sum()
```

```python
# removing rows with null values
survey = survey[survey['MoneyPerMonth'].notnull()].copy()
survey.shape
```

```python
# how many null values in country column
survey['CountryLive'].isnull().sum()
```

```python
# removing rows with null values
survey = survey[survey['CountryLive'].notnull()].copy()
survey.shape
```

```python
# creating new df only with 4 top countries
countries = ['United States of America', 'India', 'United Kingdom', 'Canada']
survey_top4 = survey[survey['CountryLive'].isin(countries)]

# grouping data by 'CountryLive' column
survey_top4.groupby('CountryLive').mean()['MoneyPerMonth']
```

```python
colors = ['lightskyblue', 'pink', 'lightgreen', 'gold']
survey_top4.groupby('CountryLive').mean()['MoneyPerMonth'].plot.pie(
    figsize = (6,6), autopct = '%.1f%%', colors = colors)
plt.title('Money Spent for Learning', fontsize = 16)
plt.ylabel('')
plt.show()
```

The results for the United Kingdom and Canada are surprisingly low relative to the values we see for India. 

It might be that there is not enough representative data for the United Kingdom, Canada, and India, or there are some outliers (maybe coming from wrong survey answers) making the mean too big for India, or too low for the UK and Canada. Or it might be that the results are correct.

## Dealing with Extreme Outliers
Let's generate four box plots to visualize for each country (the US, India, the United Kingdom, Canada) the distribution of the variable that describes how much money each participant had spent per month.

```python
sns.set(style="whitegrid")
sns.boxplot(x='CountryLive', y='MoneyPerMonth', data=survey_top4)
plt.title('Money Spent Spent per Month per Country', fontsize = 16)
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Money Per Month, USD', fontsize = 13)
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

It's hard to spot outliers for India, Canada or the United Kingdom but there is obviously something really wrong for the United States: two persons spend each month \\$50000 or more for learning. It seems extremely unlikely, so I will remove every value that goes over \\$20,000 per month.

```python
survey_top4 = survey_top4[survey_top4['MoneyPerMonth'] < 20000]
```

```python
# recomputing the mean values
survey_top4.groupby('CountryLive').mean()['MoneyPerMonth']
```

```python
sns.boxplot(x='CountryLive', y='MoneyPerMonth', data=survey_top4)
plt.title('Money Spent Spent per Month per Country', fontsize = 16)
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Money Per Month, USD', fontsize = 13)
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

There are some extreme values left but we do not know if these students attented any expensive bootcamp or it is just wrong data. Let's start with respondents living in India. I will isolate outlies and examine their answers to other questions in the survey to figure out whether these big expenses with learning are justified:

```python
india_outliers = survey_top4[(survey_top4['MoneyPerMonth'] > 2500) & 
                             (survey_top4['CountryLive'] == 'India')]
india_outliers
```

It seems that neither participant attended a bootcamp. Overall, it's really hard to figure out from the data whether these persons really spent that much money with learning. They might have misunderstood the question and included their university tution. It seems safer to remove these two rows.

```python
survey_top4 = survey_top4.drop(india_outliers.index)
```

Now, I will examine US outliers:

```python
us_outliers = survey_top4[(survey_top4['MoneyPerMonth'] > 6000) & 
                          (survey_top4['CountryLive'] == 'United States of America')]
us_outliers
```

Out of these 11 extreme outliers, six people attended bootcamps, which justify the large sums of money spent on learning. For the other five, it's hard to figure out from the data where they could have spent that much money on learning. 

Also, the data shows that eight respondents had been programming for no more than three months when they completed the survey. They most likely paid a large sum of money for a bootcamp that was going to last for several months, so the amount of money spent per month is unrealistic and should be significantly lower (because they probably didn't spend anything for the next couple of months after the survey).

As a consequence, I will remove respondents that:

- Didn't attend bootcamps.
- Had been programming for three months or less when at the time they completed the survey.

```python
# Remove the respondents who didn't attendent a bootcamp
no_bootcamp = survey_top4[
    (survey_top4['CountryLive'] == 'United States of America') & 
    (survey_top4['MoneyPerMonth'] >= 6000) &
    (survey_top4['AttendedBootcamp'] == 0)
]

survey_top4 = survey_top4.drop(no_bootcamp.index)


# Remove the respondents that had been programming for less than 3 months
less_than_3_months = survey_top4[
    (survey_top4['CountryLive'] == 'United States of America') & 
    (survey_top4['MoneyPerMonth'] >= 6000) &
    (survey_top4['MonthsProgramming'] <= 3)
]

survey_top4 = survey_top4.drop(less_than_3_months.index)
```

And finally, let's examine Canada's outliers:

```python
canada_outliers = survey_top4[
    (survey_top4['CountryLive'] == 'Canada') & 
    (survey_top4['MoneyPerMonth'] > 4500)]

canada_outliers
```

This participant had been programming for no more than two months when he completed the survey. He seems to have paid a large sum of money in the beginning to enroll in a bootcamp, and then he probably didn't spend anything for the next couple of months after the survey. I will take the same approach here as for the US and remove this outlier.

```python
survey_top4 = survey_top4.drop(canada_outliers.index)
```

```python
# recomputing the mean values
survey_top4.groupby('CountryLive').mean()['MoneyPerMonth']
```

```python
sns.boxplot(x='CountryLive', y='MoneyPerMonth', data=survey_top4)
plt.title('Money Spent Spent per Month per Country', fontsize = 16)
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Money Per Month, USD', fontsize = 13)
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

## Choosing the Two Best Markets
Considering the results so far, one country we should definitely advertise in is the US. There are a lot of new coders living there and they are willing to pay a good amount of money each month (\$142 on average per month).

Choosing the second one is a bit more tricky. It could be eigher Canada or India. 

Canada seems to be the best second choice because people there are willing to pay roughly \\$93 per month, compared to India (\\$66) and the United Kingdom (\\$45). But taking into account the quantity of potential clients, India seems like more attractive market. There are almost twice as more potential customers in India than there are in Canada:

```python
survey_top4['CountryLive'].value_counts(normalize = True) * 100
```

At this point, it seems that there are several good options:

- Advertise in the US, India, and Canada by splitting the advertisement budget in various combinations:

    - 60% for the US, 25% for India, 15% for Canada.
    - 50% for the US, 30% for India, 20% for Canada; etc.

- Advertise only in the US and India, or the US and Canada. Again, it makes sense to split the advertisement budget unequally. For instance:

    - 70% for the US, and 30% for India.
    - 65% for the US, and 35% for Canada; etc.

- Advertise only in the US.

At this point, it's probably best to send this analysis to the marketing team and let them use their domain knowledge to decide. They might want to do some extra surveys in India and Canada and then get back for analyzing the new survey data.
